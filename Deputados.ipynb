#Aula de Pandas

### Gasto dos Deputados Federais

```
import pandas as pd
from pandas.api.types import is_string_dtype
import numpy as np
```

```
df_2018 = pd.read_csv('Ano-2018.csv', sep=';', parse_dates=['datEmissao'])
```

```
df_2018.tail().T
```

Filtro por deputados

```
df_2018 = df_2018[pd.isna(df_2018['sgPartido']) == False]
df_2018
```

```
gasto = df_2018.groupby(['txNomeParlamentar', 'sgPartido'])['vlrDocumento'].sum().sorte_values(ascending=False)
gasto
```

#Gasto por Partido
###Quantos deputados tem em cada partido ?

```
df_2018['txNomeParlamentar'].value_counts()
```

```
df_2018.groupby(['sgPartido'])['txNomeParlamentar'].nunique()
```

```
partidos = df_2018.groupby(['sgPartido'])['txNomeParlamentar'].nunique()
```

```
gasto = df_2018.groupby(['sgPartido'])['vlrDocumento'].sum()
```

```
gasto = df_2018.groupby(['sgPartido'])['vlrDocumento'].sum() 
```

```
df = pd.concat([gasto, partidos], axis=1)
df["gastosDeputados"] = df["vlrDocumento"] / df["txNomeParlamentar"]
df.sort_values(by=['gastosDeputados'], ascending=False)
```
